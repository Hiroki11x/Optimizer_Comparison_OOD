```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import wandb
import seaborn as sns
```

```python

def wandb2pd(exp_runs):
    df = pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)
    summary_df = pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)
    config_df = pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)
    name_df = pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)

    summary = [] 
    config = [] 
    name = [] 
    for exp in exp_runs: 
        summary.append(exp.summary._json_dict) 
        config.append({k:v for k,v in exp.config.items() if not k.startswith('_')}) 
        name.append(exp.name)       

    summary_df = pd.DataFrame.from_records(summary) 
    config_df = pd.DataFrame.from_records(config) 
    name_df = pd.DataFrame({'name': name}) 
    df = pd.concat([name_df, config_df, summary_df], axis=1)
    return df
```

# Civil Comments

```python
algorithm_list = ['ERM', 'IRM']
dataset_list = ['WILDS_civilcomments']
optimizer_list = ['vanilla_sgd', 'momentum_sgd', 'nesterov_momentum_sgd', 'rmsprop', 'adam']
optimizer_name_list = ['SGD', 'Momentum SGD', 'Nesterov Momentum SGD', 'RMSProp', 'Adam']
colorlist = ["#377eb8", "#ff7f00", "#4daf4a", "darkred", "#984ea3", "y", "k", "w"]
metrics = ['val_eval/acc_avg', 'test_eval/acc_wg']
metric_names = ['val-acc_avg', 'test-acc_wg']

threshold = 0 * 2
```

```python

for algorithm in algorithm_list:

    text_list = [
        f'{algorithm} SGD',
        f'{algorithm} Momentum SGD',
        f'{algorithm} Nesterov Momentum SGD',
        f'{algorithm} RMSProp',
        f'{algorithm} Adam'
    ]

    for dataset_idx, dataset in enumerate(dataset_list):

        path_list = [
            f'entity_name/project_vanilla_sgd',
            f'entity_name/project_momentum_sgd',
            f'entity_name/project_nesterov_momentum_sgd',
            f'entity_name/project_rmsprop',
            f'entity_name/project_adam'
        ]

        box_plot_list = []
        
        for i in range(len(path_list)):

            p = path_list[i]
            
            spea = []
            kend = []

            api = wandb.Api()
            exp_runs = api.runs(
                path=p,
                filters={'state':'finished'}
                )

            exp_runs
            df = wandb2pd(exp_runs)
            
            df = df[df[metrics[0]]>threshold]
            df['gap'] = df[metrics[0]]-df[metrics[1]]
            
            print(f"{optimizer_name_list[i]} / {metrics[0]}: {df[metrics[0]].mean()} / {metrics[1]}: {df[metrics[1]].mean()}")
            box_plot_list.append(df)

        df_boxplot = pd.concat(box_plot_list)

        ax = sns.boxplot(x='optimizer_name', y=metrics[0], data=df_boxplot)
        ax.set_xticklabels(['SGD', 'Momentum', 'Nesterov', 'RMSProp', 'Adam'])
        plt.ylim(0, 1.)
        plt.title(f"{dataset_list[dataset_idx]}: {algorithm}")
        plt.legend(bbox_to_anchor=(1.05, 0), borderaxespad=0, loc='lower left', fontsize=10).get_frame().set_linewidth(0)
        plt.grid(linewidth=1)
        plt.savefig(f'figs/box/filtered/boxplot_{metric_names[0]}_{algorithm}_{dataset}.pdf')
        plt.show()

        ax = sns.boxplot(x='optimizer_name', y=metrics[1], data=df_boxplot)
        ax.set_xticklabels(['SGD', 'Momentum', 'Nesterov', 'RMSProp', 'Adam'])
        plt.ylim(0, 1.)
        plt.title(f"{dataset_list[dataset_idx]}: {algorithm}")
        plt.legend(bbox_to_anchor=(1.05, 0), borderaxespad=0, loc='lower left', fontsize=10).get_frame().set_linewidth(0)
        plt.grid(linewidth=1)
        plt.savefig(f'figs/box/filtered/boxplot_{metric_names[1]}_{algorithm}_{dataset}.pdf')
        plt.show()

        ax = sns.boxplot(x='optimizer_name', y='gap', data=df_boxplot)
        ax.set_xticklabels(['SGD', 'Momentum', 'Nesterov', 'RMSProp', 'Adam'])
        plt.ylim(0, 1.)
        plt.title(f"{dataset_list[dataset_idx]}: {algorithm}")
        plt.legend(bbox_to_anchor=(1.05, 0), borderaxespad=0, loc='lower left', fontsize=10).get_frame().set_linewidth(0)
        plt.grid(linewidth=1)
        plt.savefig(f'figs/box/filtered/boxplot_gap_{algorithm}_{dataset}.pdf')
        plt.show()
```
